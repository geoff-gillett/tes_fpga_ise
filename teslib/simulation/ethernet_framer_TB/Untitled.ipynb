```python
import tes.simulation.data as sim
import numpy as np
```

```python
bytestream=sim.Data.fromfile('bytestream',np.dtype([('data', np.uint32),('last',np.int32)]),
                             'teslib','ethernet_framer_TB','vivado')
```

```python
ps=sim.PacketStream(bytestream)
```

```python
ps.packets
```

```python
for p in ps.packets:
    if p.length != len(p.payload)+24:
        print('frame:{:} len:{:} payload:{:} prot:{:}'.format(
            p.frame_sequence,p.length,len(p.payload)+24,p.protocol_sequence))
```

```python
last_seq=-1
for p in ps.packets:
    if p.frame_sequence != last_seq+1:
        print('sequence error:{:}'.format(last_seq+1))
    last_seq=p.frame_sequence
```
